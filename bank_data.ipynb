```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/API_SP.POP.TOTL_DS2_en_csv_v2_6298256.csv', skiprows=4)
```

```python
df.head()
```

```python
selected_year = '2010'
```

```python
df_sorted = df[['Country Name', selected_year]].dropna().nlargest(10, selected_year)
```

```python
plt.figure(figsize=(12, 8))
plt.bar(df_sorted['Country Name'], df_sorted[selected_year], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Total Population')
plt.title(f'Top 10 Countries by Total Population in {selected_year}')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
